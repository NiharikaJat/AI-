# Ploting graph

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('data.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
col=list(data.columns)
```

```python
col
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
for i in col:
  plt.hist(data[i])
  plt.xlabel(i)
  plt.ylabel('count')
  plt.show()
```

```python
for i in col:
  sns.boxplot(data[i])
  plt.xlabel(i)
  plt.ylabel('count')
  plt.show()
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in col:
  data[i]=le.fit_transform(data[i])
```
